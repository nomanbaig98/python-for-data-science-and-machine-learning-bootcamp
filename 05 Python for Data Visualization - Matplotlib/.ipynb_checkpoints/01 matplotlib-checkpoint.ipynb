# Matplotlib 1

```python
#matplotlib.org/gallery
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#We have two ways of plotting charts one is through Functional method and second is Object oriented
```

```python
#functional
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title comes here')
```

```python
#plt.subplot(row,column,position)
plt.subplot(1,2,1)
plt.plot(x,y,'r')
         
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented

fig = plt.figure()

#axes = fig.add_axes([left,bottom,width,height])

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Write something here')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.4,0.2,0.4,0.3])
```

# Matplotlib 2

```python
#subplot using Object oriented
```

```python
fig,axes = plt.subplots()
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)

#this arranges the graphs automatically
plt.tight_layout()
```

```python
#as axes are array type then they can surely be iterated
axes
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

# Figure size and DPI

```python
#fig = plt.figure(figsize=(width,height),dpi=100)

fig = plt.figure(figsize=(3,2),dpi=100)
```

```python
fig = plt.figure(figsize=(3,2),dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)

ax.set_title('Title')
ax.set_xlabel('X')
ax.set_ylabel('Y')


```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X squared')
ax.plot(x,x**3,label='X cubed')

#legend position
# 0 means best
#ax.legend(loc=1)
ax.legend(loc=0)
```

# Matplotlib 3 
# Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='red',lw=4,alpha=0.5,ls='-.',marker='x',markersize=20,markerfacecolor='black',markeredgewidth=3,markeredgecolor='brown')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='red',lw=4,alpha=0.5,ls='-.')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
