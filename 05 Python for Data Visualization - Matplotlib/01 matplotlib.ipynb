```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#We have two ways of plotting charts one is through Functional method and second is Object oriented
```

```python
#functional
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title comes here')
```

```python
#plt.subplot(row,column,position)
plt.subplot(1,2,1)
plt.plot(x,y,'r')
         
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented

fig = plt.figure()

#axes = fig.add_axes([left,bottom,width,height])

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Write something here')
```
